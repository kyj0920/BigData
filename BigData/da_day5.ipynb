```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])

df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
```

```python
df1
```

```python
df2
```

* 내부조인 열을 기준으로 병합(고객번호), 양쪽 모두 키가 존재하는 데이터만 연결

```python
pd.merge(df1,df2)
```

* 외부조인(outer join): 방법 2개
    >1) 키 값이 한쪽에만 있어도 데이터를 병합
    
    >2) left/right outer join으로 구분

```python
# 1)

# pd.merge(df1,df2,how="inner") #디폴트:inner조인
pd.merge(df1,df2,how="outer") # 없는 데이터값은 NaN으로 처리됨
```

```python
# 2)-1 left: 왼쪽 데이터프레임(dtf) 기준: 왼쪽 데이터 프레임에 있는 값을 기준으로 오른쪽을 뽑아서 연결함
pd.merge(df1,df2,how="left") #(1008은 없기때문에 안나옴)
```

```python
# 2)-1 right:
pd.merge(df1,df2,how="right") #(1001,1005,1006,1008)
```

```python
df1 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns=['품종', '꽃잎길이'])

df2 = pd.DataFrame({
    '품종': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]},
    columns=['품종', '꽃잎너비'])
```

```python
print(df1)
print("="*50)
print(df2)
```

```python
# 1) inner
pd.merge(df1,df2) #공통으로 가진것만 출력 

# 키 값이 동일한 데이터 여러개 있는 경우에는 모든 경우의 수를 조합해서 병합
```

```python
# 2)-1 outer left
pd.merge(df1,df2,how="left") 
```

```python
# 2)-2 outer right
pd.merge(df1,df2,how="right") 
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})

df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
```

```python
pd.merge(df1,df2)
# 컬럼의 이름이 같은 2개(고객명, 데이터)의 조건이 만족해야 함
# 고객명은 일치하지만 데이터의 경우 숫자와, 글자로 다름
# 이런 경우에는 결과값이 없음
```

```python
pd.merge(df1,df2,on="고객명")
# 2개의 컬럼 중 뽑고 싶은 기준을 on=" "에 입력하여 출력한다
# 컬럼의 이름 데이터가 2개이기때문에 데이터_x, 데이터_y로 자동 수정됨
```
