2. 바차트(막대그래프): x축을 카테고리, bar(세로), barh(가로)

```python
y=[2, 3, 1]
x=np.arange(len(y))# y길이 3 = range(3)= (0,1,2)

xlabel=['가', '나', '다'] # 0, 1, 2에 가, 나, 다를 부여
plt.title('바 차트')

plt.bar(x,y)

#ticks: 아래 눈금별 자료형 입력
plt.xticks(x, xlabel)
# plt.yticks(sorted(y)) #y정렬이기때문에 1,2,3...
plt.yticks(y)
```

```python
people=['길동','꺽정','순신','세종']
y_pos=np.arange(len(people)) # arange(4) = 0,1,2,3

# np.random.rand(2) 0~1사이, 2개의 난수 발생
performance=3+10*np.random.rand(len(people)) # 0~1사이, 4개의 난수 
# performance=3+10*error 위랑 같음
error=np.random.rand(len(people))

print(performance)
plt.barh(y_pos, performance, xerr=error, alpha=0.7) # x축, y축, 에러바

# ,xerr=error: 오차범위 나타낼때 맨뒤에 표기
# alpha=0~1: 투명도 (투명~불투명)
```

4. 파이차트: 데이터를 비교할 때 많이 사용되는 그래프

```python
labels=['개구리', '돼지', '개', '통나무']
sizes=[15,30,45,10]
ex=(0,0.1,0,0)
colors=['yellowgreen','gold','lightskyblue','lightcoral']

plt.pie(sizes, labels=labels, autopct='%.1f%%', explode=ex, startangle=90, shadow=True)
# ,autopct='%.1f%%': 소숫점 첫째자리까지 입력하고, 그 이하는 반올림
# %%:뒤에 %기호 1개 출력
# ,explode= "원차트 내 띄우고 싶은 데이터에 숫자값": 숫자값만큼 떨어져 나옴
# ,startangle=90: 첫번째 값인 개구리가 90도 방향에서 시작해서 반시계방향으로 돌겠다
# ,shadow=True: 원형차트에 그림자 생성
```

* scikit-learn 함수: 머신러닝
> 과학기술과 관련된 다양한 함수들이 제공되는 패키지, 머신러닝에서 굉장히 많이 사용되는 패키지

```python
from sklearn.datasets import load_digits

digits=load_digits()
x=digits.images[0] #1797 = [[ ]] 쌍의 개수 
# 숫자 0을 숫자로 나타냄
    # digits.images[1000]
    # 대괄호와 똑같은 숫자가 저장되어 있음(1의 자리숫자만 보면됨)

plt.imshow(x,cmap=plt.cm.bone_r)
# imshow: 이미지 데이터를 시각화(plt에 있는 함수:plt.imshow)
# ,cmap=plt.cm.(tap:스타일선택) / bone_r을 많이 씀

plt.imshow(x,cmap=plt.cm.bone_r, interpolation='hamming')
# ,interpolation='gaussian': 픽셀처럼 보이지 않도록 흐리게 해줌 (+ hamming, nearest ...)
```

```python
# 입력변수가 2개, 출력변수가 1개 => 등고선(contour) : 3차원 데이터를 시각화
# 등고선 출력 함수 : contour(등고선), contourf(색상)

def f(x,y):
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2)
x=np.linspace(-3,3,256) # -3 ~ 3 사잇값을 256개로 쪼갬 / [-3,-2.97,....., 3]
y=np.linspace(-3,3,256)
xx,yy=np.meshgrid(x,y) # 매쉬그리드 행렬 생성, 입력데이터 (값을 맞춰줌)
zz=f(xx,yy) # 출력데이터
plt.contourf(xx,yy,zz, cmap='jet', alpha=0.75)
#, contourf: cmap='jet' 색상
plt.contour(xx,yy,zz,colors='black')
#, contour: 


# print(xx.shape)
# print("="*50)
# print(yy)
```

```python
# x=[1,3]
# y=[2,4,6] 
x=np.linspace(-10,10,101)
y=np.linspace(-10,10,101)
X,Y=np.meshgrid(x,y)  # print(X.shape) (101,101)  # meshgrid: shape을 맞춰주는 역할을 함
# print(X)
# print("="*50)
# print(Y)

Z=np.sqrt(X**2+Y**2) #sqrt 제곱근
# Z=X*Y #요소간 곱셈 (주의: 행렬 곱셈 아님)
# 행렬의 곱셈 : [3,2][2,1]=>[3,1]
# print("="*50)
# print(Z)
# 행렬의 곱셈, 덧셈하려면 앞의 열,행이 같아야함

cs=plt.contourf(X,Y,Z,100) #100은 레벨임
plt.colorbar(cs)
# plt.contour(X,Y,Z,colors="black") #colors=black은 등고선
```
