```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'cust_id': ['c1', 'c1', 'c1', 'c2', 'c2', 'c2', 'c3', 'c3', 'c3'], 
'prod_cd': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
'grade' : ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'], 
'pch_amt': [30, 10, 0, 40, 15, 30, 0, 0, 10]})
```

```python
data
```

* pivot() / pivot_table()
> * pivot_table()을 주로 사용(pivot()은 오류 발생 가능성이 높음)
> * 인덱스가 2개인 경우 pivot_table()은 정상수행, pivot은 에러
> * 컬럼이 2개인 경우 pivot_table은 정상 수행, pivot은 에러


```python
# pandas > dataframe > pivot = data.pivot
data_pivot=data.pivot(index="cust_id", columns="prod_cd", values="pch_amt")
data_pivot
```

```python
# pandas > pivot_table = pd.pivot_table
pd.pivot_table(data, index="cust_id", columns="prod_cd", values="pch_amt")
```

```python
# 인덱스가 2개인 경우 pivot은 에러
# data.pivot(index=["cust_id",'grade'], columns="prod_cd", values="pch_amt")

# 인덱스가 2개인 경우 pivot_table은 정상 수행
# pd.pivot_table(data, index=['cust_id','grade'], columns="prod_cd", values="pch_amt")

# -------------------------------------------------------------------

# 컬럼이 2개인 경우 pivot은 에러
# data.pivot(index='cust_id', columns=['grade','prod_cd'], values="pch_amt")

# 컬럼이 2개인 경우 pivot_table은 정상 수행
pd.pivot_table(data, index='cust_id', columns=['grade','prod_cd'], values="pch_amt") 
#데이터가 없던 부분은 NaN으로 자동 처리됨
```

```python
data
```

```python
# pivot 함수는 중복값이 있는 경우 에러가 발생
# data.pivot(index='grade', columns='prod_cd', values='pch_amt')

# 디폴트로 평균 함수가 적용됨
# pd.pivot_table(data, index='grade', columns='prod_cd', values='pch_amt', aggfunc=np.mean, margins=True)
pd.pivot_table(data, index='grade', columns='prod_cd', values='pch_amt', aggfunc=np.sum, margins=True)
# aggfunc= (집계함수): 데이터를 계산하는 함수
# sum = margins=True > All이라는 컬럼과 인덱스가 생성되며, 합계를 계산해줌
# mean = margins=True > 합계의 평균
```

```python
# 여러 개의 열을 하나로 정리하기
weather=pd.read_csv("data/weather.csv")
weather


# 날짜 열을 행 데이터로 피벗
weather_melt=pd.melt(weather, id_vars=['id','year','month','element'], var_name="day", value_name="temp")
weather_melt


# pivot_table 함수 이용하여 아래와 같은 형식으로 출력
#                                   tmax tmin
# 'id','year','month','element'

#1), 2) 둘다 같음

#1)
wether_tidy=weather_melt.pivot_table(index=["id","year","month","day"], 
                           columns= "element", 
                           values= "temp" )

#2)
wether_tidy=pd.pivot_table(weather_melt, 
                           index=["id","year","month","day"], 
                           columns= "element", 
                           values= "temp" )
wether_tidy

#행 인덱스들이 모두 데이터로 변환됨
wether_tidy_flat=wether_tidy.reset_index()
wether_tidy_flat
```
