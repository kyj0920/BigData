# 각 컬럼별 누락값의 개수를 반환하는 함수

```python
import numpy as np
import pandas as pd

def count_missing(vec):
    null_vec = pd.isnull(vec) #True(NaN)/False(NaN이 아닌값)
    null_count = np.sum(null_vec)
#     print("="*50)
    return null_count   
#     return 각 컬럼별 누락값의 개수
cmis_col=titanic.apply(count_missing) 
print(cmis_col)

#axis를 생략했으므로, 기본값인 0(열 방향)
```

```python
x=[1,2,3,4,5,np.nan]
pd.isnull(x) # array([False, False, False, False, False,  True])
np.sum(pd.isnull(x)) #false는0. true는 1
```

```python
# 퍼센트 구하기

def prop_missing(vec):
    num=count_missing(vec)
    return num/vec.size # size 데이터의 크기를 알수 있음
#     return 각 컬럼별 nan의 비율


def prop_complete(vec): #null이 아닌 값들의 비율을 리턴해주는 식
    return 1-prop_missing(vec)  # 1. 함수호출, 2. 1에서 값 빼기, 3. prop_complet에 넣기
```

```python
pmis_col=titanic.apply(prop_missing)
print(pmis_col)

pmis_col=titanic.apply(prop_complete)
print(pmis_col)
```

```python
cmis_row=titanic.apply(count_missing, axis=1)  #행단위로 nan 확인하겠음
pmis_row=titanic.apply(prop_missing, axis=1) 
pcom_row=titanic.apply(prop_complete, axis=1) 

# cmis_row
# titanic.loc[5]
pmis_row
pcom_row
```

```python
#각 행 단위로 누락값의 개수('num_missing'를 구하여 타이타닉 데이터프레임에 추가)
titanic['num_missing']=titanic.apply(count_missing, axis=1)
titanic
```

# 3. 표준점수
- 데이터의 평균과 표준편차의 차이 
- 표준점수를 구하면 변환된 데이터의 평균값이 0, 표준편차 1이 됨
- 데이터가 표준화 되어 있어야 서로 다른 데이터를 쉽게 비교할 수 있게 됨

```python
def my_zscore(x): 
    return (x - x.mean()) / x.std() #mean:6.2, std:3.7
#분자는 연산순서를 고려하여 반드시 묶어야 함
# mean함수는 array에 적용가능하지만, list에는 적용 불가
# mean, std함수는 넘파이에서 제공되는 함수
# 넘파이에서는 기본적으로 함수 적용 대상이 array(배열)

data=[1,3,7,9,11]
data=np.array(data) # 파이썬list -> 넘파이 array로 바뀜                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      
my_zscore(data)
```

```python
x=[1,3,7,9,11]
sum(x)/5 #평균: 6.2
np.std(x) #표준편차: 3.7
np.var(x) #분산: 13.76
```
