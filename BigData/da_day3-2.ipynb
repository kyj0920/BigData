```python
import seaborn as sns
```

```python
anscombe=sns.load_dataset("anscombe")
```

```python
anscombe
# 엔스콤 데이터는 데이터 그룹 1,2,3,4 모두 평균, 분산, 상관관계 같다
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
dataset_1 = anscombe[anscombe["dataset"] == 'I']
dataset_2 = anscombe[anscombe["dataset"] == 'II']
dataset_3 = anscombe[anscombe["dataset"] == 'III']
dataset_4 = anscombe[anscombe["dataset"] == 'IV']
dataset_1
dataset_2
dataset_3
dataset_4
# 데이터그룹 1을 추출하여 저장
```

```python
# # 선 그래프 : plot 메서드
# # plot 메서드에 x,y축 데이터를 전달 -> 선 그래프 출력
# plt.plot(dataset_1['x'],dataset_1['y'])

plt.plot(dataset_1['x'],dataset_1['y'], 'o')
#'o'를 입력하면 기존 출력값에 데이터 부분을 점으로 나타냄
#'o' 옵션은 동그란 점으로 데이터를 출력
```

```python
#   시각화 과정
#  1) 전체 그래프가 출력될 기본 틀(figure) 생성
fig=plt.figure()
#  2) 틀을 그래프 격자로 나눔
axes1=fig.add_subplot(2,2,1)# 틀을 2x2로 나누고, 1번째 위치(왼쪽 상단부터 시작)를 axes1
axes2=fig.add_subplot(2,2,2)
axes3=fig.add_subplot(2,2,3)
axes4=fig.add_subplot(2,2,4)
#  3) 각 격자에 그래프를 추가
axes1.plot(dataset_1['x'],dataset_1['y'], 'o')
axes2.plot(dataset_2['x'],dataset_2['y'], 'o')
axes3.plot(dataset_3['x'],dataset_3['y'], 'o')
axes4.plot(dataset_4['x'],dataset_4['y'], 'o')
#  4) 각 격자에 그래프 출력


axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

fig.suptitle("Anscombe Data")
fig.tight_layout() #데이터 안겹쳐지게 함
```

```python
# 시각화도구 : matplotlib.seaborn
tips=sns.load_dataset("tips") #밥먹고 팁얼마주는지
tips

# print(tips['total_bill'].max())
# print(tips['total_bill'].min())
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1) #axis(축), axes 한칸을 만들어서 한칸을 다 쓰겠다
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Frequency")
axes1.set_title("Histogram of Total Bill")

axes1.hist(tips["total_bill"],bins=20)

#total bill에 대한 빈도수
# bins=20)
```

```python
#산점도 그래프: 변수 2개로 그린 그래프, 이변량 그래프라고도 함
# 식사 비용에 따른 팁 열의 분포를 산점도 그래프로 나타내보자
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tips")
axes1.set_title("Scratterplot of Total Bill vs Tip")
#이상치 찾기
```

변수2 : 이산형 변수(명확하게 구분되는 값), 연속형 변수(명확하게 셀 수 없는 범위의 값)

boxplot(상자그림, 이산형 변수와 연속형 변수를 함께 사용하는 그래프)

```python
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
             tips[tips['sex']=='Male']['tip']],
             labels=["Female", 'Male'])
# boxplot([x축의 첫번째에 출력데이터, x축의 두번째 출력데이터], 라벨은 각 데이터의 이름을 부여)
# 동그란 원 하나가 1개임(여성을 기준으로 팁을 굉장히 많이 준 내역이 1건 있음)
```

```python
tips[tips['sex']=='Female']
```

```python
# 다변량 그래프 그리기: 변수가 3개 이상 있을때
# 데이터 점의 색상을 달리 할 수 있음
```

```python
def recode_sex(sex):
    if sex=="Female":
        return 0
    else:
        return 1
```

```python
tips['sex_color']=tips['sex'].apply(recode_sex)

# 작성형식 : 데이터.apply(적용함수)
# apply: 특정 데이터에 대해 특정 함수를 적용하고자 할때 사용
```

```python
tips
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
axes1.set_title("Total Bill vs Tip - 30")
axes1.scatter(x=tips['total_bill'],
              y=tips['tip'],
              c=tips['sex_color'],
              alpha=0.8 )
# alpha는 투명도임
```
