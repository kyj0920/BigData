```python
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline
```

```python
import seaborn as sns
=sns.load_dataset("tips")
```

```python
tips.head()
```

-----------------------------------------------------------------------------------------------------------------------------------------
* seaborn 히스토그램: subplots(기본 틀), distplot(데이터 전달)메서드를 사용하여 시각화

```python
ax=plt.subplots()
# ax=sns.distplot(tips['total_bill'], hist=False)
# 히스토그램 + 밀집도 그래프(선) 함께 출력
# , kde=False : 밀집도 그래프 제외해라
# , hist=False : 히스토그램 제외해라

ax=sns.distplot(tips['total_bill'], rug=True)
# rug(양탄자)그래프: 가로 일직선으로 밀집도를 나타냄
```

```python
ax=plt.subplots()
ax=sns.countplot('day', data=tips) 
#countplot
#('기준 index', data=데이터이름)
# day 기준, 빈도수 (토, 일, 목, 금 순서)
ax.set_title("Count of days")
tips
```

```python
# 산점도 그래프 + 회귀선
sns.regplot(x="total_bill", y="tip", data=tips, fit_reg=False)
# , fit_reg=False : 회귀선 미출력
```

------------------------------------------------------------------------------------------------------------
* jointplot 함수: 산점도 그래프 + 히스토그램

```python
ax=sns.jointplot(x='total_bill',y='tip', data=tips) 

# 피어슨 상관계수(= pearsonr)
# -1 ~ 1: +-1에 가까울 수록 상관관계가 높다. 0에 가까울수록 상관관계가 낮다
# (+1)양의 상관관계: 우상향 / (-1)음의 상관관계: 우하향
```

```python
ax=sns.jointplot(x='total_bill',y='tip', data=tips, kind='hex') 
# ,kind='hex': 육각형 모형으로 출력하여 겹쳐진 산점도 그래프의 구분이 가능하도록 함
```

-----------------------------------------------------------------------------------------------------------------------------------------
* kdeplot 함수: 이차원 밀집도 그리기

```python
ax=sns.kdeplot(data=tips['total_bill'],data2=tips['tip'], shade="True") 
# shade를 안넣으면 등고선 형태로 나옴
# 좌측 아래부분의 total_bill이 집중되어 있음
```

------------------------------------------------------------------------------------------------------------
* 시간(Lunch, Dinner)에 따라 지불한 비용의 평균을 바 그래프로 출력

```python
ax=sns.barplot(x="time", y="total_bill", data=tips)
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

-----------------------------------------------------------------------------------------------------------------------------------------
* boxplot: 최소값, 이상치 등 다양한 정보를 한꺼번에 제공하는 그래프

```python
ax=sns.boxplot(x="time", y="total_bill", data=tips)
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

-----------------------------------------------------------------------------------------------------------------------------------------
* violinplot: 큰쪽의 데이터가 많음

```python
ax=sns.violinplot(x="time", y="total_bill", data=tips)
ax.set_title("title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
```

-----------------------------------------------------------------
* 관계 그래프: pairplot메서드
> 우하향 대각선을 기준으로 아래쪽 절반과 위쪽 절반이 같음(중복데이터)

```python
fig=sns.pairplot(tips) 
# 3*3 그래프: 수치데이터를 가진 인덱스가 3가지
# ex) tip은 독립변수, total_bill 종속변수: tip에 따라 total_bill이 바뀜
# 중복값을 지우기 위어느한쪽을 공간으로 나타내야함
```

```python
pair_grid=sns.PairGrid(tips)
pair_grid.map_upper(sns.regplot)
pair_grid.map_lower(sns.kdeplot)
pair_grid.map_diag(sns.distplot, rug=True)
```

```python
new_row_series
```

```python
pd.concat([df1, new_row_series])
# 데이터프레임 + 시리즈 = 시리즈가 새로운 열로 추가, 나머지 NaN
# 시리즈가 추가되면서 0 인덱스가 자동으로 부여되어 데이터가 출력됨
```

* 데이터프레임에 새로운 행으로 시리즈를 연결하려고하면 연결이 되지 않음
