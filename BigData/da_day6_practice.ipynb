# 데이터 표준점수 구하기

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# 기업: CJ Logistics Corporation (000120.KS)
# 기간: Apr 15, 2018 - Apr 15, 2019
```

```python
# 데이터 불러오기 
df=pd.read_csv("Downloads/000120.KS.csv")
df.head()
```

```python
# 데이터 정보
df.info()
```

```python
# 데이터 내 컬럼별 열 추출
df_open=df.Open
df_high=df.High
df_low=df.Low
df_close=df.Close
df_adj_close=df.Adj_Close
df_volume=df.Volume
```

```python
# 컬럼 array로 변경
df_open=np.array(df_open)
df_high=np.array(df_high)
df_low=np.array(df_low)
df_close=np.array(df_close)
df_adj_close=np.array(df_adj_close)
df_volume=np.array(df_volume)
```

```python
# 표준화 함수 작성
def my_zscore(x): 
    return (x - x.mean()) / x.std()
```

```python
# 표준화 함수 대입 & 기존데이터 대입
df.Open=my_zscore(df_open)
df.High=my_zscore(df_high)
df.Low=my_zscore(df_low)
df.Close=my_zscore(df_close)
df.Adj_Close=my_zscore(df_adj_close)
df.Volume=my_zscore(df_volume)
```
