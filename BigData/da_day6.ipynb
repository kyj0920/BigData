```python
# ignore

tips_sub_miss['total_bill']=pd.to_numeric(
    tips_sub_miss["total_bill"], errors='ignore') 
# 오류들을 무시해라(여전히 total_bill은 문자열로)
# 숫자로 된 문자인 경우에는 to_numeric 사용시 숫자로 변환됨(missing 문자가 숫자와 함께 입력되어 있음)
tips_sub_miss.dtypes
```

```python
# coerce

tips_sub_miss['total_bill']=pd.to_numeric(
    tips_sub_miss["total_bill"], errors='coerce') 
# coerce: 강제로 누락값으로 바꿔라 (total_bill은 float로 바뀜)
tips_sub_miss.dtypes
tips_sub_miss['total_bill']
```

```python
#문자열 관련 메서드
x="good morning"
x.capitalize() #첫 글자가 대문자로
x.count('n') # 변수에 글자 n의 개수를 세는 함수

# 어떤 문자로 시작하는지 확인하고 싶을 때
x.startswith('Good') #False
x.startswith('good') #True

# 어떤 문자로 끝나는지 확인하고 싶을 때
x.endswith('ing')

x.find('n') # n글자의 첫번째 인덱스 리턴: 8
x.find('z') # -1출력 = 없음

x.index('n') # n글자의 첫번째 인덱스 리턴: 8
x.index('z') # 없으므로 에러 발생(find와의 차이점)
```

```python
# 알파벳(문자) 확인하기
x.isalpha() # 공백이 포함되어 있어 false
"good".isalpha() # True
"good2".isalpha() # 숫자가 포함되어 있어서 false

"1234".isdecimal() # 모든 문자가 숫자면 True

"a1234".isalnum() # 숫자와 알파벳으로 구성되어있으면 True
"1234".isalnum() # True
"good".isalnum() # True
"good morning".isalnum() # 공백문자가 있으므로 false

"good morning".upper() #대문자변환
"GOOD MORNING".lower() #소문자변환
```

```python
"good morning".replace("o", "k") # o문자가 모두 k로 바뀜
"good morning".replace("oo", "ab") #oo문자가 ab로 바뀜
```

```python
import re
tele_num="1234567890"
re.match("\d+",tele_num)
```

```python
# 전화번호를 패턴으로 저장
tel_num_spaces="123 456 7890" 
m=re.match("\d{10}", tel_num_spaces) #공백인식 못해서 매치안됨
print(m)
if m:
    print("매치됨")
else:
    print("매치 안됨")
```

```python
tel_num_spaces="123 456 7890"
m=re.match("\d{3}\s\d{3}\s\d{4}", tel_num_spaces) #3자리+공백 3자리+공백 +4자리
print(m)
```

```python
# compile을 이용한 매치여부 확인
pat=re.compile("\d{10}")
s="1234567890"
m=pat.match(s)
print(m)
```
