```python
# 문자열 형식 -> 날짜 형식으로 변환
# test_df1['date_dt1']=pd.to_datetime(test_df1['order_day'])
# default 형식: 월/일/년도 -> 년/월/일

test_df1['date_dt1']=pd.to_datetime(test_df1['order_day'],format='%m/%d/%y')
# format 형식으로 만들어진 데이터임                              #01/01/15
# print(test_df1)
# test_df1.info()

test_df1['date_dt2']=pd.to_datetime(test_df1['order_day'],format='%d/%m/%y')
# print(test_df1)

test_df1['date_dt3']=pd.to_datetime(test_df1['order_day'],format='%y/%m/%d')
print(test_df1)
```

```python
# 일, 월, 년 = Yy, m, d
# 년도의 자릿수에 따라 Y는 대/소문자를 맞춰줘야 함
# Y: 년도 4자리, y:년도 2자리
print(pd.to_datetime('2019/04/16',format='%Y/%m/%d'))
print(pd.to_datetime('19/04/16',format='%y/%m/%d'))

# 시, 분, 초 = H, M, S (대문자로 작성해야함)
print(pd.to_datetime('19/04/16/09',format='%y/%m/%d/%H'))
print(pd.to_datetime('19/04/16/09/54',format='%y/%m/%d/%H/%M'))
print(pd.to_datetime('19/04/16/09/54/50',format='%y/%m/%d/%H/%M/%S'))
```

```python
ebola=pd.read_csv("data/country_timeseries.csv")
ebola['date_dt']=pd.to_datetime(ebola['Date'])
ebola[['Date','date_dt']].head()
```

```python
#3번 행 인덱스의 date_dt 컬럼에 저장된 연도를 추출
print(ebola['date_dt'][3].year)
print(ebola['date_dt'][3].month)
print(ebola['date_dt'][3].day)
# 위에 기술된 코드는 인덱스로 특정 데이터를 접근해야 하기때문에 

# dt 접근자를 이용하여 쉽게 접근이 가능함
# dt라는 접근자(속성을)는 한번에 접근해서 연을 뽑아냄
ebola['year']=ebola['date_dt'].dt.year
ebola.info()
ebola[['Date','date_dt','year']]  #.head()
```

```python
# 월, 일 데이터를 한 번에 추출(dt이용)해서 새로운 열(month, day)로 추가
# 1) = 2)

# 1)
# ebola['month']=ebola['date_dt'].dt.month
# ebola['day']=ebola['date_dt'].dt.day

# 2)
ebola['month'],ebola['day']=ebola['date_dt'].dt.month, ebola['date_dt'].dt.day

ebola.info()
```

에볼라 최초 발병일 계산

```python
# ebola.iloc[-5:] # 하위 5개
ebola.iloc[-5:,:5] # 하위 5개부터 왼쪽부터 5번째 칼럼
ebola.tail()
```

```python
# min, max 적용하기 위해서는 수치함수여야 함
# date_dt컬럼값이 날짜(수치)이므로 min사용이 가능
ebola['date_dt'].min() #최소값 
ebola['date_dt'].max() #최대값
```

```python
# Date 컬럼은 object이기 때문에 수치연산이 불가함

# 발병한지 얼마나 지났는지 알아봄
ebola['outbreak_d']=ebola['date_dt']-ebola['date_dt'].min()
ebola[['Date','Day','outbreak_d']].head()
ebola[['Date','Day','outbreak_d']].info()
```

```python
# 파산한 은행의 개수 계산
banks=pd.read_csv("data/banklist.csv")
banks.head()
banks.info()
```

```python
banks=pd.read_csv("data/banklist.csv", parse_dates=[5,6])
banks.head()
banks.info()
```

```python
# 분기, 연도 뽑기
banks['Closing_quarter'], banks['Closing_year'] = banks['Closing Date'].dt.quarter, banks['Closing Date'].dt.year
print(banks.head())
```

```python
# 연도별로 파산한 은행이 몇개인지 출력
closing_year=banks.groupby(['Closing_year']).size() # banks에서 groupby / 숫자세기는 size()
closing_year
```

```python
# 연도별로 그룹화한 다음 다시 분기별로 파산한 은행이 몇개인지 출력
closing_year_q=banks.groupby(['Closing_year','Closing_quarter']).size() # banks에서 groupby / 숫자세기는 size()
closing_year_q
```

```python
import matplotlib.pyplot as plt

# 년도별 파산한 은행수
fig,ax = plt.subplots()
ax=closing_year.plot()
plt.show()
```

```python
fig,ax = plt.subplots()

# 분기별 파산한 은행수
ax=closing_year_q.plot()
plt.show()
```
