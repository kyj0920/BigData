```python
# 기술통계: 사분위수, 중앙값, 최대/최소, 표준편차...
x=np.array([5,1,3,2,4,8,7,50])
len(x)
np.mean(x) # 10
print(np.var(x)) # 분산 -> n으로 나눔
print(np.sqrt(np.var(x))) # 표준편차
print(np.std(x)) # 표준편차
print(np.max(x)) # 최대값
print(np.min(x)) # 최소값
print(np.median(x)) # 중앙값
#평균이 10인데 중앙값이 4.5인것을 보면 극단적으로 큰 값이 존재해서 왜곡이 발생함
print(np.percentile(x,0)) #최소값
print(np.percentile(x,25)) #1사분위
print(np.percentile(x,50)) #2사분위
print(np.percentile(x,75)) #3사분위
print(np.percentile(x,100)) #4사분위

# 데이터 처리 전에 수치 전체 확인하고 시작
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits

# 사이킷 런 패키지의 datasets 모듈에 있는 load_digits 함수를 가져와라
digits=load_digits()
samples=[0,10,20,30,1,11,21,31]
d=[]
digits 

# digits['data'][0].reshape(8,8)
# 숫자가 작으면 옅은색, 숫자가 크면 진한색
# 숫자 '0'임

for i in range(8):
    d.append(digits.images[samples[i]])
    
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r) #숫자 데이터를 이미지로 변환하여 출력
    plt.title("img {}".format(i+1))
    
plt.tight_layout()
plt.show()

# len(digits['data']) #1797개의 숫자가 들어있음
# digits['data'][10] #
# digits['target_names'][digits['target'][10]]
```

```python
#위에서 아래로 길게 출력됨

v=[]
# np.shape(d) #(8,8,8) : 8x8이미지가 8개
for i in range(8):
    v.append(d[i].reshape(64,1)) # 8*8 > 64*1
plt.figure(figsize=(8,3)) #이미지 크기
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(v[i], interpolation='nearest', cmap=plt.cm.bone_r) #숫자 데이터를 이미지로 변환하여 출력
    plt.title("img {}".format(i+1))
    
plt.tight_layout(w_pad=10)
plt.show()
```

```python
x=np.array([[1],[2],[3]]) #3*1
y=np.array([[4],[5],[6]]) #3*1
```

```python
# np.dot() : 행렬의 곱셈
np.shape(x) #(3,1)
np.shape(y) #(3,1)

np.shape(x.T) #(1, 3) > 전치행렬
res=np.dot(x.T,y) #(1, 3)*(3,1)=(1*1)   >  array([[32]])
np.shape(res)# 2차원  > (1, 1)
res[0][0] # 32 출력됨

# np.dot(x,y) # 1 (dim 1) != 3 (dim 0) > 1, 3이 같지 않아서 안됨(에러)
```
