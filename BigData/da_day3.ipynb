pandas => Datetime 자료형: 날짜를 나타내는 수치형 (-를 빼기가 아닌 날짜로 인식함)

```python
born_datetime=pd.to_datetime(scientists['Born'])
# ~으로 변경해라 '.to_' / pd의 sientists 내 born 데이터를 datetime으로 변경해라
```

```python
born_datetime
```

```python
born_datetime=pd.to_datetime(scientists['Born'],format="%Y-%m-%d")
born_datetime
# 작성방법을 알고 싶으면 http://pandas.pydata.org/의 pdf 파일에 ctr F + 함수값을 입력해라
# format="%Y-%m-%d" 포맷을 지정해주면 적용됨. datetime 사용하는 경우 디폴트 값으로 들어 있어서 안적어도 됨
died_datetime=pd.to_datetime(scientists['Died'],format="%Y-%m-%d")
died_datetime
#"%Y-%m-%d" (Y(대문자):연도 4자리, y(소문자):연도 2자리)
# 시, 초, 분 = %H:%M:%S'
```

```python
scientists['born_dt'], scientists['died_dt']=born_datetime, died_datetime
```

```python
scientists.shape
# 7개 컬럼
```

```python
# scientists
# scientists['Born']-'scientists['Died'] 문자끼리는 뺄셈이 안된다.
scientists['age_days_dt']=scientists['died_dt']-scientists['born_dt']
```
