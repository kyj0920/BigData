누락값(NaN): NaN, NAN, nan 등으로 일반적인 표기, 누락값을 처리하기 위해 넘파이가 필요함 
* 바구니 조차 없고, 존재하지 않는 값

```python
from numpy import NaN,NAN,nan
import pandas as pd

# 3가지 Nan을 다 데려오겠음
```

```python
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == " ") # 바구니는 있는데 그안에 내용물이 없음
print(NaN == NaN) # 값이 아니므로 비교 자체를 수행 못함
# NaN 자체는 특정 값이 아님
```

```python
# isnull: 누락값을 확인하는 메서드 
pd.isnull(NaN)
x=[1,2,3,NaN]
pd.isnull(x)
```

```python
# notnull: null이 아니다
pd.notnull(10) #10은 누락값이 아니므로 true
pd.notnull("test") #누락값이 아니므로 true
pd.notnull(x)
```

```python
ebola=pd.read_csv("data/country_timeseries.csv")
ebola.index
ebola.columns
ebola.info()
```

```python
ebola.count()#누락값이 아닌 개수
ebola.shape # (122, 18) => 행, 렬
ebola.shape[0]-ebola.count() # 누락값의 개수(NaN의 개수)
```

```python
import numpy as np

ebola.isnull()

np.count_nonzero(ebola.isnull()) #nonzero : 0(False)이 아닌것
# count_nonzero:0(False)가 아닌 값들의 개수를 세는 함수

cnt_nan=122*18-np.count_nonzero(ebola.isnull())
np.count_nonzero(ebola.isnull())

# ebola.isnull():null이면 True, 아니면 False
# count_nonzero:0이 아닌 것의 개수 1(True)의 개수
```

```python
ebola['Cases_Guinea'].count() # 122 중 nan이 아닌 개수: 93건
```

```python
ebola['Cases_Guinea'].isnull() #nan이면 true
np.count_nonzero(ebola['Cases_Guinea'].isnull()) #True(1)의 개수가 몇개임?
```

```python
ebola.Cases_Guinea
ebola.Cases_Guinea.value_counts() #각각의 값들의 개수(Nan은 제외)
ebola.Cases_Guinea.value_counts(dropna=True) #dropna>nan을 제거한 나머지
#NaN은 제외, 디폴트값이
ebola.Cases_Guinea.value_counts(dropna=False) #nan 포함
ebola.Cases_Guinea.value_counts(dropna=False).head() #상위 다섯개 뽑겠음
```

```python
# fillna(): 특정값으로 na를 대체
ebola.fillna(0) # nan을 0으로 대체
ebola.fillna(0).iloc[0:10, 0:5] # 10행 5열로 출력


```

```python
# ebola
ebola.fillna(method="ffill")
# 순방향(위에서 아래로)으로 이동해가면서 NaN을 만나게 되면,
# 직전에(위) 있는 데이터 값으로 NaN을 대체
```

```python
ebola.fillna(method="bfill")
# 바로 밑에 있는 데이터 값으로 NaN을 대체
```
