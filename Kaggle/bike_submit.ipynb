```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
train=pd.read_csv("data/train_bike.csv", parse_dates=["datetime"])
train.shape
train.info()
```

```python
test=pd.read_csv("data/test_bike.csv", parse_dates=["datetime"])
test.shape
test.info()
```

```python
train['year']=train['datetime'].dt.year
train['month']=train['datetime'].dt.month
train['hour']=train['datetime'].dt.hour
train['dayofweek']=train['datetime'].dt.dayofweek
train.shape
train.info()
```

```python
test['year']=test['datetime'].dt.year
test['month']=test['datetime'].dt.month
test['hour']=test['datetime'].dt.hour
test['dayofweek']=test['datetime'].dt.dayofweek
test.shape
test.info()
```

```python
# 범주형(category)
train['season']=train['season'].astype('category')
test['season']=test['season'].astype('category')

fn=['season', 'holiday', 'workingday', 'weather']
for f in fn:
    train[f]=train[f].astype('category')
    test[f]=test[f].astype('category')
train.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_wind=RandomForestRegressor()
fcolumns=['season', 'weather', 'humidity', 'month', 'temp', 'year', 'hour', 'dayofweek', 'atemp']
xtrain=train[fcolumns]
xtrain
yrlabel=train['registered']
yclabel=train['casual']

# rfModel=RandomForestRegressor() 
# rfModel.fit()
```

```python
yrlabel
```

```python
xtest=test[fcolumns]
xtest.shape
```

```python
rfmodel=RandomForestRegressor()
rfmodel.fit(xtrain, yrlabel)   # train['registered']
# xtrain 데이터 변수가 9개 변수로 구성되어있음(x1~x9, y)
# rfmodel.fit(x1~x9, y) => yhat=w1x1 + w2x2 + ... + w9x9 + b
# yhat은 예상되는 registered 값
test['resgistered']=rfmodel.predict(xtest)
test['resgistered'].isnull().sum()
```

```python
rfmodel.fit(xtrain, yclabel)  
test['casual']=rfmodel.predict(xtest)
test['casual'].isnull().sum()
```
