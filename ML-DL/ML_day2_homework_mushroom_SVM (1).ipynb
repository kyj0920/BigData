```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
import random
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import LabelEncoder 
```

```python
mushroom=pd.read_csv('agaricus-lepiota.csv',encoding='utf-8')
mushroom
```

```python
from sklearn.preprocessing import LabelEncoder #문자를 수치로 변환해줌 
le=LabelEncoder()
for col in mushroom.columns:
    mushroom[col]=le.fit_transform(mushroom[col])
mushroom #1이 독버섯 0이 일반버섯
```

```python
df_rank=np.abs(mushroom.corr().classes).sort_values(ascending=False)
m_all=mushroom[df_rank.index[0:9]]
m_all
m_label=m_all.classes
m_label
m_data=m_all.iloc[:,1:]
m_data
```

```python
data_train,data_test,label_train,label_test=train_test_split(m_data,m_label)
```

```python
len(data_train)
len(data_test)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(data_train,label_train)
```

```python
prediction_SVM_all=classifier.predict(data_test)
```

```python
metrics.accuracy_score(label_test,prediction_SVM_all)
```

```python
rep=metrics.classification_report(label_test,prediction_SVM_all)
print(rep)
```
