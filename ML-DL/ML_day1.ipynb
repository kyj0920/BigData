```python
data - mean(data, axis=0) #data 의 각 열데이터에서 평균값을 뺀 값이 산출됨
```

```python
std(data, axis=0) # 각 열의 표준편차
```

```python
data_std_np=(data - mean(data, axis=0)) / std(data, axis=0) # 컬럼(열) 단위로 각각의 평균, 표준편차를 적용한 표준화
```

```python
mean(data_std_np, axis=0) # 각 열의 평균 
std(data_std_np, axis=0) # 각 열의 표준편차
```

```python
# 범주형 변수가 주어졌을때 이항변수화(OneHotEncoder) #하나만 핫하게 만드는 인코더
# 성별: 남=>1, 여=>2, 연령: 10대 =>1, 20대 =>2...
# 등급: S=>1, A=>2...
# ex)
# 성별(남(0)/여(1)) => 원핫인코딩 => 
# 0=>10, 1=>01
# 연령대(10대(0),20대(1),30대(2),40대(3))
# 0=>1000, 1=>0100, 2=>0010, 3=>0001
```

```python
from sklearn.preprocessing import OneHotEncoder
data=np.array(
#[성별, 연령대, 등급]
[[0,0,0],
 [0,1,1],
 [0,2,2],
 [1,0,3],
 [1,1,4]])
```

```python
data
```

```python
ohe=OneHotEncoder()
ohe.fit(data) #data를 원핫인코더(ohe)를 이용하여 변환(fit)한다
```

```python
ohe.active_features_
#[남/여, 20/30/40, S/A/B/C/D] => 10개의 종류가 있다
```

```python
ohe.n_values_
#성별:2, 연령:3, 등급:5  범주의 개수를 출력
```

```python
# 여성, 30대, S등급
mydata=np.array([[1,1,0]])
ohe.transform(mydata).toarray()
#1=>01, 1=>01, 0=>10000 (2개, 3개, 5개)
ohe.feature_indices_#array([ 0,  2,  5, 10], dtype=int32): 0이상 2미만(성별), 2이상 5미만(연령), 5이상 10미만(등급)
```
